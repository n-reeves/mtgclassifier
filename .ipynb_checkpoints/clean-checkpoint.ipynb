```python
import pandas as pd
import re
raw_data = pd.read_csv('test3.csv')
```

```python
def type_func(type):
    typestr = type.lower()
    if "land" in typestr:
        final_type = "Land"
    elif "creature" in typestr:
        final_type = "Creature"
    elif "enchantment" in typestr:
        final_type = "Enchantment"
    elif "sorcery" in typestr:
        final_type = "Sorcery"
    elif "instant" in typestr:
        final_type = "Instant"
    elif "artifact" in typestr:
        final_type = "Artifact"
    else:
        final_type = "Other"
    return final_type
```

```python
raw_data['type2'] = raw_data['type'].apply(lambda x: type_func(x))
```

```python
def color_func(color):
    color_str = color
    color_str = color.replace("[","").replace("]", "").replace("'", "").replace(", ","")
    return color_str
```

```python
raw_data['color_identity'] = raw_data['color_identity'].apply(lambda x: color_func(x))
```

```python
def text_clean(text):
    new_text = re.sub(r'[.()\n,]', ' ', str(text).lower())
    new_text = re.sub(r'[\']', '', new_text)
    new_text = re.sub(r'first strike', 'firststrike', new_text)
    new_text = re.sub(r'double strike', 'doublestrike', new_text)
    return new_text
```

```python
data['text'] = data['text'].apply(lambda x: text_clean(x))
```

```python
def to_artifact(color):
    if color is '':
        return 'A'
    else:
        return color
```
